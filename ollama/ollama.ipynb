Generate Function

```python
#Generation
response = ollama.generate(model = "llama3.2:1b" , prompt = "why does moon glow in one line")
print(response)
print(response.response)
```

```python
#Stream Response
response = ollama.generate(model = "llama3.2:1b" , prompt = "why does moon glow?", stream= True) # stream = True ka mtlb jaise jaise output generate hoga vaise vaise response aayi jayega
for i in response :
    print(i['response'], end ="")
```

```python
# Giving multimodel input to model ( passing image to the model )
import base64 # jiski help se mei encoding kar rha hu
import ollama
image_path = "download.webp"
with open ( image_path = 'rb') as f: # opening the image in read binary
    image_bytes = f.read()
image_64 = base64.b64encode(image_bytes).decode('utf-8') # encoding the image in base 64 bits
response = ollama.generate(model = 'gemma3:4b', images=[image_64],prompt ="Give caption for the image") # model must support images and videos and is model ke pass vision capability haii
print(response.response)
```

```python
#passing multiple images to the model
images_path = ['download.webp', 'download1.webp']
images_base64 = []
for i in images_path:
    with open ( image_path = 'rb') as f: # opening the image in read binary
        image_bytes = f.read()
        images_base64.append(base64.b64encode(image_bytes).decode('utf-8'))
response = ollama.generate(model = 'gemma3:4b', images=images_base64,prompt ="Generate a story based on these images, make sure you take context from each and every image")
print(response.response)
```

```python
#System Instructions
response = ollama.generate(model='llama3.2:1b', prompt = " why does moon glow in 30 words.", system="You are a emotional ai assistant , you explain each in emotional way")
response.response
```

```python
#Options parameter
# The options dictionary contains generation parameters passed to the model.
# temperature Controls creativity/randomness.
# top_p Controls probability cutoff.
# top_k Model only considers top 45 most probable next tokens.
response = ollama.generate(model = 'llama3.2:1b',prompt="Why is the ocean blue",
                           options = {
                               'temperature':0.3,
                               'top_p':0.5,
                               'top_k':45
                           }
                           )
print(response.response)
```
