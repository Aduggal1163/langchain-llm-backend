```python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_community.tools import tool
from langgraph.prebuilt import create_react_agent
from langchain_ollama import ChatOllama
import requests
```

```python
search_result = DuckDuckGoSearchRun()
```

```python
llm = ChatOllama(
    model= 'llama3.2:1b',
    temperature = 0
)
```

```python
@tool
def get_weather(city : str)->str:
    """
    This function finds the current weather for a particular city
    """
    url = f'https://api.weatherstack.com/current?access_key=0078760375d955ed5fa7d9ac3ed85c88&query={city}'
    response = requests.get(url)
    return response.json()
```

```python
agent = create_react_agent(
    model = llm,
    tools = [search_result,get_weather]
)
```

```python
response = agent.invoke({
    'messages':[
        ('user',"Find the capital of punjab, India and also find temperature in C")
    ]
})
```
