```python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_ollama import ChatOllama
search_result = DuckDuckGoSearchRun()
llm = ChatOllama(
    model="llama3.2:1b",
    temperature=0
)
response = llm.invoke("Plan a 5 day trip to Bali , also tell the weather and also make it under 60k")
print(response.content)
```

```python
import requests
from langchain_community.tools import tool
@tool
def get_weather_details(city: str)->str:
    """
    This function generate a weather report for 5 consecutive days
    """
    url = f"http://api.weatherapi.com/v1/forecast.json?key=8d6c75fe91d548749b763632262202&q={city}&days=5"
    response = requests.get(url)
    data = response.json()
    forecast_days = data["forecast"]["forecastday"]
    result = f"5-Day Weather for {city}:\n"
    for day in forecast_days:
        result += f"{day['date']} - {day['day']['condition']['text']} - {day['day']['avgtemp_c']}°C\n"
    return result
```

```python
print(get_weather_details.invoke({"city": "Bali"}))
```

```python
@tool
def currency_converter(amount : float)->float:
    """
    This function converts INR to IDR
    """
    url = f'https://api.exchangerate-api.com/v4/latest/INR'
    response = requests.get(url)
    data = response.json()
    converte_rates = data['rates']['IDR']
    converted_money = amount*converte_rates
    return converted_money
```

```python
print(currency_converter.invoke({"amount": 1000.00}))
```

```python
from langgraph.prebuilt  import create_react_agent
agent = create_react_agent(
    model = llm,
    tools = [search_result,get_weather_details,currency_converter]
)
response = agent.invoke({
    'messages':[
        ("user", """
            Plan a 7-day trip to Goa, India.
Budget must be under ₹30,000.

Follow these steps strictly:

1. First, call the get_weather_details tool for "Goa".
2. Print the 7-day weather clearly (date-wise).
3. Then create a 7-day trip plan.
4. For each day include:
   - Activities
   - Estimated cost for that day
   - Weather for that specific date
5. At the end calculate total cost properly.

At the end write strictly in this format:
TOTAL_COST: <number only>

Do not skip weather tool.
            """)
    ]
})
print(response['messages'][-1].content)
```
