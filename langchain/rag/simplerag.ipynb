```python
#directly reading from web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
#load chunk and index the content of HTML page or any web page
loader = WebBaseLoader(web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
                       bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                           class_=("post-title","post-content","post-header")
                       )),)
text_documents=loader.load()
```

```python
text_documents
```

```python
#pdf reader
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("plan.pdf")
docs=loader.load()
docs
```

```python
for page in docs:
    print(page.page_content)
```

TRANSFORM (large data to small chunks)


```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
documents=text_splitter.split_documents(docs)
documents[:5]

```

```python
for docs in documents:
    print(docs.page_content)
    print("----------")
```

NOW CONVERT IT INTO VECTORS

```python
#vector embedding techniques(convert text into vectors) and vector store ( after creating vectors store it into the vector store)
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
# create embedding model (IMPORTANT)
embeddings = GoogleGenerativeAIEmbeddings(
    model="text-embedding-004"
)
db=Chroma.from_documents(documents,embeddings)
# now this db can be stored in local or cloud anywhere 
```

```python
#vector database
query = "who are the authors of the attention is all you need research papers"
result = db.similarity_search(query)
result
```
